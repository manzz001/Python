```python
import random
import string
```

```python
def generate_password():
    length = int(input('Enter the length of the password : ').strip())
    include_uppercase = input('Include upper case (yes/no) : ').lower().strip()
    include_special = input('Include specialo charecters (yes/no) : ').lower().strip()
    include_digits = input('Include digts (yes/no) : ').lower().strip()
    
    if length < 4:
        print('Password length must at least 4 charecters.')
        return
        
    
    lowercase = string.ascii_lowercase
    uppercase = string.ascii_uppercase if include_uppercase == 'yes' else ""
    special = string.punctuation if include_special == 'yes' else ""
    digits  = string.digits if include_digits == 'yes' else ""
    all_charecters = lowercase + uppercase + special + digits

    required_charecters = []
    if include_uppercase == 'yes':
        required_charecters.append(random.choice(uppercase))
    if include_special == 'yes':
        required_charecters.append(random.choice(special))
    if include_digits == 'yes':
        required_charecters.append(random.choice(digits))

    remaining_length = length - len(required_charecters)
    password = required_charecters

    for _ in range(remaining_length):
        charecter =  random.choice(all_charecters)
        password.append(charecter)

    random.shuffle(password)

    str_password = "".join(password)
    return str_password
        

password = generate_password()
print(password)

    
    
```
