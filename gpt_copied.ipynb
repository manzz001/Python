```python
import json

# Initialize the to-do list
to_do_list = []

# Load existing tasks from file (if any)
def load_tasks(file_name='to_do_list.json'):
    global to_do_list
    try:
        with open(file_name, 'r') as file:
            to_do_list = json.load(file)
            print("âœ… Tasks loaded successfully.")
    except FileNotFoundError:
        print("ğŸ“‚ No existing file found. Starting fresh.")

# Save tasks to file
def save_tasks(file_name='to_do_list.json'):
    with open(file_name, 'w') as file:
        json.dump(to_do_list, file, indent=4)
        print("ğŸ’¾ Tasks saved successfully.")

# View tasks
def view_tasks():
    if not to_do_list:
        print("ğŸ“­ No tasks available.")
    else:
        print("ğŸ“‹ Current Tasks:")
        for idx, task in enumerate(to_do_list, start=1):
            print(f"{idx}. {task}")

# Add a new task
def add_task():
    task = input("ğŸ“ Enter the new task: ")
    to_do_list.append(task)
    print("âœ… Task added.")

# Update an existing task
def update_task():
    view_tasks()
    try:
        task_num = int(input("ğŸ”¢ Enter task number to update: "))
        if 1 <= task_num <= len(to_do_list):
            new_task = input("âœï¸ Enter the updated task: ")
            to_do_list[task_num - 1] = new_task
            print("âœ… Task updated.")
        else:
            print("âŒ Invalid task number.")
    except ValueError:
        print("âš ï¸ Please enter a valid number.")

# Main menu loop
def main():
    load_tasks()
    while True:
        print("\nğŸ“Œ To-Do List Manager")
        print("1. View Tasks")
        print("2. Add Task")
        print("3. Update Task")
        print("4. Save Tasks")
        print("5. Create New List")
        print("6. Exit")

        choice = input("ğŸ‘‰ Choose an option (1-6): ")

        if choice == '1':
            view_tasks()
        elif choice == '2':
            add_task()
        elif choice == '3':
            update_task()
        elif choice == '4':
            save_tasks()
        elif choice == '5':
            to_do_list.clear()
            print("ğŸ†• New list created.")
        elif choice == '6':
            print("ğŸ‘‹ Exiting. Goodbye!")
            break
        else:
            print("âš ï¸ Invalid choice. Try again.")

# Run the program
main()
```
